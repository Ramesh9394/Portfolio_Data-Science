# Exploratory Data Analysis
### Data Cleaning

```python
# Check for datatypes
data.info()
```

- **Need to Change datatyps:**
    - last_review - datetime    

```python
# Change the datatype 
data['last_review'] = pd.to_datetime(data['last_review'])
```

```python
# Verifying the change in datatype
data.info()
```

```python
# id and host_name are irrelevaent and insignificant for our analysis, so will drop both variables.
data.drop(["id","host_name"], axis=1, inplace=True)
```

```python
# Checking for missing values
data.isnull().sum()
```

```python
# Percent missing values
data.isnull().sum() * 100 / data.shape[0]
```

```python
# Lets check if any pattern in the missing data
data[data.isna().any(axis=1)]
```

`last_review` is date feature if there is no reviews for that month, then there will be missing last_review date. So we can drop this column. For `reviews_per_month` as no reviews for that month, we camn fill in missing values with '0.0'.

```python
# Drop last_review
data.drop("last_review", axis=1, inplace=True)
```

```python
# Fill missing values in reviews_per_month by 0.0
data['reviews_per_month'].fillna(0.0, inplace=True)
```
